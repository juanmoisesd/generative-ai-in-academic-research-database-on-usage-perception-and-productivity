Analysis of generative AI tool adoption, perception, and productivity impact in Latin American academic research contexts.

````python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, warnings
warnings.filterwarnings('ignore'); sns.set_theme(style='whitegrid'); print('Libraries loaded')
````

````python
import os
df=None
for p in ['../data/genai_academic_research.csv','../data/data.csv']:
    if os.path.exists(p): df=pd.read_csv(p); break
if df is None:
    df=pd.DataFrame({'tool':['ChatGPT','Copilot','Gemini','Perplexity','Claude','Midjourney'],'adoption_pct':[78.3,42.1,38.5,29.7,21.4,15.2],'trust_score':[3.8,3.5,3.4,3.6,4.0,3.2],'productivity_gain_pct':[35.2,28.4,24.1,20.3,31.5,18.7],'discipline':['All','STEM','All','All','SS','Arts']})
    print('Using sample data')
print(df.shape); df.head()
````

````python
fig,axes=plt.subplots(1,2,figsize=(13,5))
df.sort_values('adoption_pct',ascending=False).plot.bar(x='tool',y='adoption_pct',ax=axes[0],legend=False,color='steelblue')
axes[0].set_title('AI Tool Adoption Rate (%)')
df.sort_values('productivity_gain_pct',ascending=False).plot.bar(x='tool',y='productivity_gain_pct',ax=axes[1],legend=False,color='green')
axes[1].set_title('Reported Productivity Gain (%)')
for ax in axes: ax.tick_params(axis='x',rotation=45)
plt.tight_layout(); plt.show()
````
